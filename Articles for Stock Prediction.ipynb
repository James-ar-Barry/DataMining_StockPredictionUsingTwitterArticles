```python
import pandas as pd
col_names = ['Article','Date']
df = pd.read_csv('Articles.csv', index_col=False)

df
```

```python
import re
from __future__ import division
i=0

threshold=0
dates = []
pos_sentiment = []
neg_sentiment = []
neut_sentiment = []
Number_of_articles = []

for k in range(3,5):
    for j in range(1,32):
        pos_sent_article=0
        neg_sent_article=0
        null_pol = 0
        print(j)

        if j < 9: 
            start_date = ('2017-0%s-0%s'%(k, j))
            end_date = ('2017-0%s-0%s'%(k,(j+1)))
        elif j == 9 :
            start_date = ('2017-0%s-0%s'%(k, j))
            end_date = ('2017-0%s-%s'%(k,(j+1)))
        else:
            start_date = ('2017-0%s-%s'%(k, j))
            end_date = ('2017-0%s-%s'%(k,(j+1)))

        dates.append(start_date)

        mask = (df['Date'] >= start_date) & (df['Date'] < end_date)
        df1 = df.loc[mask]
        Number_of_articles.append(len(df1['Article']))
        Number_of_articles_in = len(df1['Article'])
        for article in df1['Article']:
            if article.isalnum:
                analysis=TextBlob(article)
        #         print(i)
        #         i = i+1
        #         print(article)
        #         print(analysis.sentiment.polarity)
                if analysis.sentiment.polarity>threshold:
                    pos_sent_article=pos_sent_article+1
                elif analysis.sentiment.polarity<threshold:
                    neg_sent_article=neg_sent_article+1
                else:
                    null_pol = null_pol + 1
        print('The articles collected between: %s and %s'%(start_date, end_date))
        print('Number of positive articles are: %d' % pos_sent_article)
        print('Number of negative articles are: %d' % neg_sent_article)
        print('Number of neutral articles are: %d' % null_pol)
        if Number_of_articles_in == 0:
            pos_sentiment.append(1)
            neg_sentiment.append(0)
            neut_sentiment.append(0)
        else:
            pos_sentiment.append(pos_sent_article/Number_of_articles_in)
            neg_sentiment.append(neg_sent_article/Number_of_articles_in)
            neut_sentiment.append(null_pol/Number_of_articles_in)

#     if pos_sent_article>neg_sent_article:
#         print("The Overall sentiment is: Positive")
#         sentiment.append(1)
#     else:
#         print ("The Overall sentiment is: Negative")
#         sentiment.append(0)


```

```python

# df3 = pd.DataFrame({'Date': dates, 'Sentiment': sentiment, 'Number_of_Articles': Number_of_articles})
df3 = pd.DataFrame({'Date': dates, 
                    'Pos_A_Sentiment': pos_sentiment,
                    'Neg_A_Sentiment': neg_sentiment, 
                    'Neut_A_Sentiment': neut_sentiment,
                    'Number_of_Articles': Number_of_articles})

df3.to_csv("Articles_sentiment2.csv", index=False)

df3
```
